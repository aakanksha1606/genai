#### What is the difference between "Love" and "love" in NLP?
In the context of NLP the both given words are different due to case sensitive.
"Love" (capital L)
May represent a proper noun 
Example: Love is a beautiful thing.
"love" (lowercase)
Usually treated as a common word
Example: I love coding.
If preprocessing includes lowercasing, both become "love" and are treated the same.
If not, NLP models may treat them as different tokens.

#### What happens if stopwords are not removed?
Stopwords = common words like is, the, and, in, to
If they are not removed:
Increased data size → more memory and computation
Noise in data → less meaningful patterns
Reduced model performance 
Important words may get less weight due to frequent stopwords
However, removing them is not always good.

#### Provide two real-world scenarios where removing stopwords can be harmful.
Scenario 1: Sentiment Analysis
Sentence: "This movie is not good"
If "not" is removed:
Becomes: "This movie is good".
 Meaning completely changes (negative → positive)
Scenario 2: Question Answering / Search Queries
Query: "What is the best college in India?"
Removing stopwords:
"best college India"
Loses structure and intent.
In such cases, stopwords carry important semantic meaning

#### Explain the difference between stemming and lemmatization.
Stemming :
Cuts words to root form. Stemming cuts the word in to the root word displaying the main word. Sometimes it may not be real words.
Lemmatization :
Converts to meaningful base form. In Lemmatization the word is also cuts down but into a meaningful base word.

```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stpowords')
nltk.download('wordnet')
```

```python
stop_words= set(stopwords.words('english'))
lemmatizer= WordNetLemmatizer()
```

```python
text= [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
print(f'Original Text:{text}')
```

```python
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # 1. Lowercase
    text = text.lower()

    # 2. Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 4. Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 5. Remove special characters (keep words only)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. Tokenization
    tokens = text.split()

    # 8. Remove short tokens (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # 9. Final cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
results = []

for texts in text:
    tokens, clean_sent = preprocess_text(texts)
    results.append((texts, tokens, clean_sent))

for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Cleaned:", r[2])
```

```python
import statistics

analytics = []

for text, tokens, clean_sent in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0

    analytics.append({
        "text": text,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": avg_len
    })

for a in analytics:
    print("\nText:", a["text"])
    print("Total Tokens:", a["total_tokens"])
    print("Unique Tokens:", a["unique_tokens"])
    print("Avg Token Length:", a["avg_token_length"])
```

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[:-6:-1])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sent = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sent)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(text)
print(output)
```
